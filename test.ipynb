```python
import mysql.connector

config = {
    'user': 'root',
    'password': '18Ezequiel,',
    'host': 'localhost',
    'database': 'proyecto0012',
    'port': '3306',
}

# Conectarse a la base de datos
cnx = mysql.connector.connect(**config)

# Iniciar una transacción
cnx.start_transaction()

try:
    # Borrar todos los datos de la tabla de origen
    cursor = cnx.cursor()
    cursor.execute("DELETE FROM main")

    # Insertar los datos en la tabla de destino
    cursor.execute("INSERT INTO registro () SELECT * FROM main")

    # Confirmar la transacción
    cnx.commit()
    print("Transacción completada con éxito")

except Exception as e:
    # Si ocurre algún error, deshacer la transacción
    cnx.rollback()
    print("Error durante la transacción:", e)

finally:
    # Cerrar el cursor y la conexión a la base de datos
    cursor.close()
    cnx.close()
```
